```python
import matplotlib.pyplot as plt
import numpy as np

import kuibit.cactus_scalars as cs
import kuibit.simdir as sd

from kuibit import grid_data as gd
```

```python
sim_names = ['TOV_res_1', 'TOV_res_2', 'TOV_res_4']
```

```python
sims = []

for f in sim_names:
    sims.append(sd.SimDir(f"../simulations/{f}"))
```

```python
max_rhos = [sim.timeseries.maximum.fields.rho for sim in sims]
print(max_rhos)
```

```python
for f, rho_max in zip(sim_names,max_rhos):
    plt.plot(rho_max, label=f"{f}")
    plt.xlabel(r'$t$ [$M_{\odot}$]')
    plt.ylabel(r'$\rho_c / \rho_c(0)$')
    plt.title('NS density evolution')
    plt.ylim(1e-3, 1.4e-3)
    plt.legend()

plt.savefig(f'{f}.png', dpi=250, bbox_inches='tight')
```
