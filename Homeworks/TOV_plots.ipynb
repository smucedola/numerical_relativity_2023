```python
import matplotlib.pyplot as plt
import numpy as np

import kuibit.cactus_scalars as cs
import kuibit.simdir as sd

from kuibit import grid_data as gd
```

```python
d = "TOV_res"  #class
f = "TOV_1_lev" #name of the simulation folder
```

```python
sim = sd.SimDir(f"../simulations/{d}/{f}")
```

```python
timeseries = sim.timeseries
print(timeseries)
```

```python
rho_max=timeseries.maximum.fields.rho
```

```python
plt.plot(rho_max, label="maximum rest-mass density")
plt.xlabel(r'$t$ [$M_{\odot}$]')
plt.ylabel(r'$\rho_c / \rho_c(0)$')
plt.title('NS density evolution')
plt.legend()


plt.savefig(f'{f}.png', dpi=250, bbox_inches='tight');
```

```python
gf = sim.gf

print(gf)
```

```python
vars2D = gf.xy

print(vars2D)
```

```python
rho = vars2D.fields.rho
```

```python
print(rho.iterations)
```

```python
print(rho.available_times)
```

```python
small_grid = gd.UniformGrid([100, 100], x0=[0, 0], x1=[12,12])

rho_small = rho.read_on_grid(rho.iterations[-1], small_grid)

#cf = plt.contourf(*rho_small.coordinates_meshgrid(), np.log10(rho_small.data_xyz), vmin=-10, vmax=-3)
cf = plt.pcolormesh(*rho_small.coordinates_meshgrid(), np.log10(rho_small.data_xyz),
                    vmin=-10, vmax=-3, cmap='inferno')
plt.colorbar(cf)
```

```python
from IPython.display import HTML
import matplotlib.animation as animation
from tqdm.notebook import tqdm
ims = []
fig,ax = plt.subplots()

for i in tqdm(rho.iterations[0::1]):
    rho_small = rho.read_on_grid(i, small_grid)
    ax.set_xlabel(r'$x (M_\odot)$')
    ax.set_ylabel(r'$y (M_\odot)$')
    ax.set_aspect('equal')
    
    im = ax.pcolormesh(*rho_small.coordinates_meshgrid(), np.log10(rho_small.data_xyz), 
                       animated=True, vmin=-10, vmax=-3, cmap='inferno')
    plt.plot()
    
    title = ax.text(0.5,1.05,
                    'Time='+str(rho.available_times[i//2048]), 
                    size=plt.rcParams["axes.titlesize"],
                    ha="center", transform=ax.transAxes)
    
    ims.append([im, title])
    plt.close()
    

ani = animation.ArtistAnimation(fig, ims, interval=50, blit=True, repeat_delay=1000)
HTML(ani.to_jshtml())
```

```python
ffwriter = animation.FFMpegWriter(fps=24)
ani.save(f'{f}.mp4', writer=ffwriter, dpi = 150)
```
